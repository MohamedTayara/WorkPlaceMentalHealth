```python
# Importing needed libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data
# Define the file ID from the Google Drive link
file_id = '1QdwGmQ_reheRPOnOCfHNm0KZUeLvIvuA'

# Create the download URL
url = f'https://drive.google.com/uc?id={file_id}'

# Read the CSV file into a DataFrame
df_data = pd.read_csv(url)

# Display the first few rows of the DataFrame
df_data.head(5)
```

```python
df_data.describe()
```

```python
# Discovering the Null Values

df_data.isnull().sum()
```

```python
# All data type are fine for the purpose of the project, then we move to the next step
# Cleaning the data by replacing null values 

# Replacing null values in the 'Mental_Health_Condition' column with 'Healthy'
df_data['Mental_Health_Condition'] = df_data['Mental_Health_Condition'].fillna('Healthy')

# Replacing null values in the 'Physical_Activity' column with 'Lazy'
df_data['Physical_Activity']= df_data['Physical_Activity'].fillna('lazy')
df_data.isnull().sum()
```

```python
df_data.describe()
```

```python
# Next I am comparing the work location effect on mental health condition
#Creating crosstab comparison

work_location_mental_health = pd.crosstab(df_data['Work_Location'], df_data['Mental_Health_Condition'])
print(work_location_mental_health)
```

```python
# Plotting the crosstab
ax = work_location_mental_health.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Distribution of Mental Health Conditions by Work Location')
plt.xlabel('Work Location')
plt.ylabel('Employee Count')
plt.legend(title='Mental Health Condition', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside
plt.xticks(rotation=0)  # Keep x-tick labels horizontal

# Adding numbers on the bars
for i in range(len(work_location_mental_health)):
    cumulative_height = 0  # Initialize cumulative height for each bar
    for j in range(len(work_location_mental_health.columns)):
        height = work_location_mental_health.iloc[i, j]
        if height > 0:  # Only annotate if the height is greater than 0
            ax.text(i, cumulative_height + height / 2,  # Position the text in the middle of the segment
                    str(height), 
                    ha='center', 
                    va='center', 
                    color='black')
            cumulative_height += height  # Update cumulative height

plt.show()
```

```python

# Assuming df_data is already defined and contains the relevant columns
work_location_mental_health = pd.crosstab(df_data['Work_Location'], df_data['Mental_Health_Condition'])

# Calculate the percentage of each mental health condition for each work location
percentage_df = work_location_mental_health.div(work_location_mental_health.sum(axis=1), axis=0) * 100

# Display the percentage DataFrame and its columns
print(percentage_df)
print("Column names:", percentage_df.columns)

# Rank based on the 'Healthy' condition
ranked_work_locations = percentage_df['Healthy'].sort_values(ascending=False)
print("Ranked Work Locations based on Good Mental Health (Healthy):")
print(ranked_work_locations)
```

```python

# Calculate the percentage of each mental health condition for each work location
percentage_df = work_location_mental_health.div(work_location_mental_health.sum(axis=1), axis=0) * 100

# Rank based on the 'Healthy' condition
ranked_work_locations = percentage_df['Healthy'].sort_values(ascending=False)

# Plotting the findings
plt.figure(figsize=(10, 6))
ranked_work_locations.plot(kind='bar', color='skyblue')
plt.title('Percentage of Healthy Individuals by Work Location')
plt.xlabel('Work Location')
plt.ylabel('Percentage of Healthy Individuals (%)')
plt.xticks(rotation=0)  # Keep x-tick labels horizontal
plt.ylim(0, 100)  # Set y-axis limit from 0 to 100%
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for better readability

# Adding the percentage values on top of the bars
for index, value in enumerate(ranked_work_locations):
    plt.text(index, value + 1, f"{value:.1f}%", ha='center', va='bottom')

plt.show()
```

```python
#Access to Mental Health Resources and Productivity Change
#Some employees with mental health conditions (like "Depression" or "Anxiety") do not have access to mental health resources (e.g., see rows where "Access_to_Mental_Health_Resources" is "No").
#The "Productivity_Change" column could reflect whether access to mental health resources impacts productivity.
#Questions:
#Does access to mental health resources impact employees’ productivity positively or negatively?
#Is there a correlation between mental health conditions and productivity change?
```

```python
# Step 1: Create a crosstab to compare Productivity_Change across conditions and access to resources
crosstab_df = pd.crosstab(
    [df_data['Mental_Health_Condition'], df_data['Access_to_Mental_Health_Resources']], 
    df_data['Productivity_Change'], 
    normalize='index'
) * 100  # Convert to percentages

# Step 2: Display the results
print("\nProductivity Change Comparison by Mental Health Condition and Access to Resources:\n")
print(crosstab_df)

# Plotting
crosstab_df.plot(kind='bar', stacked=False, figsize=(12, 8), colormap='viridis', width=0.8)

# Adding labels and title
plt.title('Productivity Change Comparison by Mental Health Condition and Access to Resources')
plt.xlabel('Mental Health Condition and Access to Resources')
plt.ylabel('Percentage of Productivity Change')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.legend(title='Productivity Change')

# Display the plot
plt.show()
```
