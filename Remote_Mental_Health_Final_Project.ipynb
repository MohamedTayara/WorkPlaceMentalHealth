```python
# Importing needed libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data
# Define the file ID from the Google Drive link
file_id = '1QdwGmQ_reheRPOnOCfHNm0KZUeLvIvuA'

# Create the download URL
url = f'https://drive.google.com/uc?id={file_id}'

# Read the CSV file into a DataFrame
df_data = pd.read_csv(url)

# Display the first few rows of the DataFrame
df_data.head(5)
```

```python
# Discovering the Null Values

df_data.isnull().sum()
```

```python
# All data type are fine for the purpose of the project, then we move to the next step
# Cleaning the data by replacing null values 

# Replacing null values in the 'Mental_Health_Condition' column with 'Healthy'
df_data['Mental_Health_Condition'] = df_data['Mental_Health_Condition'].fillna('Healthy')

# Replacing null values in the 'Physical_Activity' column with 'Lazy'
df_data['Physical_Activity']= df_data['Physical_Activity'].fillna('lazy')
df_data.isnull().sum()
```

```python
# Next I am comparing the work location effect on mental health condition
#Creating crosstab comparison

work_location_mental_health = pd.crosstab(df_data['Work_Location'], df_data['Mental_Health_Condition'])
print(work_location_mental_health)
```

```python
# Plotting the crosstab
ax = work_location_mental_health.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Distribution of Mental Health Conditions by Work Location')
plt.xlabel('Work Location')
plt.ylabel('Employee Count')
plt.legend(title='Mental Health Condition', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside
plt.xticks(rotation=0)  # Keep x-tick labels horizontal

# Adding numbers on the bars
for i in range(len(work_location_mental_health)):
    cumulative_height = 0  # Initialize cumulative height for each bar
    for j in range(len(work_location_mental_health.columns)):
        height = work_location_mental_health.iloc[i, j]
        if height > 0:  # Only annotate if the height is greater than 0
            ax.text(i, cumulative_height + height / 2,  # Position the text in the middle of the segment
                    str(height), 
                    ha='center', 
                    va='center', 
                    color='black')
            cumulative_height += height  # Update cumulative height

plt.show()
```

```python
#Access to Mental Health Resources and Productivity Change
#Some employees with mental health conditions (like "Depression" or "Anxiety") do not have access to mental health resources (e.g., see rows where "Access_to_Mental_Health_Resources" is "No").
#The "Productivity_Change" column could reflect whether access to mental health resources impacts productivity.
#Questions:
#Does access to mental health resources impact employees productivity positively or negatively?
#Is there a correlation between mental health conditions and productivity change?
```

```python


# Step 1: Create a crosstab to compare Productivity_Change across conditions and access to resources
crosstab_df = pd.crosstab(
    [df_data['Mental_Health_Condition'], df_data['Access_to_Mental_Health_Resources']], 
    df_data['Productivity_Change'], 
    normalize='index'
) * 100  # Convert to percentages

# Step 2: Display the results
print("\nProductivity Change Comparison by Mental Health Condition and Access to Resources:\n")
print(crosstab_df)

# Plotting
crosstab_df.plot(kind='bar', stacked=False, figsize=(12, 8), colormap='viridis', width=0.8)

# Adding labels and title
plt.title('Productivity Change Comparison by Mental Health Condition and Access to Resources')
plt.xlabel('Mental Health Condition and Access to Resources')
plt.ylabel('Percentage of Productivity Change')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.legend(title='Productivity Change')

# Display the plot
plt.show()
```

```python
df_data_copy = df_data.copy()
df_data_copy.info()
df_data_copy.head(5)
```

```python
# Step 1: Create a copy of the DataFrame
df_data_copy = df_data.copy()

# Step 2: Clean the Stress_Level column
df_data_copy['Stress_Level'] = df_data_copy['Stress_Level'].str.strip()  # Remove any leading/trailing whitespace

# Step 3: Map stress levels to numerical values
stress_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
df_data_copy['Stress_Level'] = df_data_copy['Stress_Level'].map(stress_mapping)

# Step 4: Calculate the average stress level by work location
avg_stress_work_place = df_data_copy.groupby('Work_Location')['Stress_Level'].mean().sort_values()

# Convert Series to DataFrame for heatmap
avg_stress_df = avg_stress_work_place.reset_index()
avg_stress_df.columns = ['Work_Location', 'Average_Stress_Level']

# Step 5: Create the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(avg_stress_df.set_index('Work_Location').T, annot=True, cmap='coolwarm', fmt=".2f", cbar_kws={'label': 'Average Stress Level'})
plt.title("Average Stress Level by Work Location")
plt.xlabel("Work Location")
plt.ylabel("")

# Display the heatmap
plt.show()
```

```python
# Step 1: Count healthy individuals in every job role
healthy_count = df_data['Mental_Health_Condition'].value_counts().get('Healthy', 0)
healthy_by_job_role = df_data.groupby('Job_Role')['Mental_Health_Condition'].apply(lambda x: (x == 'Healthy').sum())

# Step 2: Count depression in every job role
depression_by_job_role = df_data.groupby('Job_Role')['Mental_Health_Condition'].apply(lambda x: (x == 'Depression').sum())

# Step 3: Count anxiety in every job role
anxiety_by_job_role = df_data.groupby('Job_Role')['Mental_Health_Condition'].apply(lambda x: (x == 'Anxiety').sum())

# Step 4: Count burnout in every job role
burnout_by_job_role = df_data.groupby('Job_Role')['Mental_Health_Condition'].apply(lambda x: (x == 'Burnout').sum())

# Step 5: Combine the results into a single DataFrame
mental_health_counts = pd.DataFrame({
    'Healthy': healthy_by_job_role,
    'Depression': depression_by_job_role,
    'Anxiety': anxiety_by_job_role,
    'Burnout': burnout_by_job_role
}).fillna(0)  # Fill NaN with 0 for job roles with no counts

# Display the results
print(mental_health_counts)
```

```python
# Grouped Bar Chart
ax = mental_health_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Mental Health Conditions by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Move the legend to the bottom
plt.legend(title='Mental Health Condition', loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=4)  # ncol=4 for a horizontal legend
plt.tight_layout()
plt.show()
```

```python
# Step 1: Count the number of healthy individuals by Job_Role and Work_Location
healthy_counts = df_data[df_data['Mental_Health_Condition'] == 'Healthy'].groupby(['Job_Role', 'Work_Location']).size().reset_index(name='Healthy_Count')

# Step 2: Find the healthiest location for each job role
healthiest_locations = healthy_counts.loc[healthy_counts.groupby('Job_Role')['Healthy_Count'].idxmax()]

# Step 3: Display the results
print(healthiest_locations)
```

```python
# PLotting the findings:

# Step 1: Plotting the healthiest locations for each job role
plt.figure(figsize=(10, 6))

# Create a horizontal bar plot
sns.barplot(data=healthiest_locations, 
             x='Healthy_Count', 
             y='Job_Role', 
             hue='Work_Location', 
             dodge=True)

# Customize the plot
plt.title('Healthiest Work Location by Job Role')
plt.xlabel('Count of Healthy Individuals')
plt.ylabel('Job Role')
plt.legend(title='Work Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Step 1: Count the number of healthy individuals by Job_Role and Work_Location
healthy_counts = df_data[df_data['Mental_Health_Condition'] == 'Healthy'].groupby(['Job_Role', 'Work_Location']).size().reset_index(name='Healthy_Count')

# Step 2: Find the healthiest location for each job role
healthiest_locations = healthy_counts.loc[healthy_counts.groupby('Job_Role')['Healthy_Count'].idxmax()]

# Step 3: Sort the healthiest locations by Healthy_Count in descending order
healthiest_locations_sorted = healthiest_locations.sort_values(by='Healthy_Count', ascending=False)

# Step 4: Plot the results

plt.figure(figsize=(12, 6))

# Create a horizontal bar plot
sns.barplot(data=healthiest_locations_sorted, 
             x='Healthy_Count', 
             y='Job_Role', 
             hue='Work_Location', 
             dodge=True)

# Customize the plot
plt.title('Healthiest Work Location by Job Role (Descending Order)')
plt.xlabel('Count of Healthy Individuals')
plt.ylabel('Job Role')
plt.legend(title='Work Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Finding out the corelation between virtual meetinfs number and stress level:

# Step 1: Group by Stress_Level and calculate the average number of virtual meetings
virtual_meetings_by_stress = df_data.groupby('Stress_Level')['Number_of_Virtual_Meetings'].mean()

# Display the results
print(virtual_meetings_by_stress)

# Step 2: Plot the relationship between Number of Virtual Meetings and Stress Level
plt.figure(figsize=(8, 6))
sns.boxplot(x='Stress_Level', y='Number_of_Virtual_Meetings', data=df_data, order=['Low', 'Medium', 'High'])
plt.title('Number of Virtual Meetings by Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Number of Virtual Meetings')

plt.show()
```

```python
# Group by Work_Location and Satisfaction_with_Remote_Work, calculate the counts, and unstack
location_satisfaction = (
    df_data.groupby(['Work_Location', 'Satisfaction_with_Remote_Work']).size()
    .unstack()  # Pivot Satisfaction_with_Remote_Work values into columns
    .div(df_data['Work_Location'].value_counts(), axis=0) * 100  # Calculate percentage across Work_Location
)

# Optionally, format with the % symbol
location_satisfaction = location_satisfaction.applymap(lambda x: f"{x:.2f}%" if pd.notnull(x) else "0.00%")

# Display the result
print(location_satisfaction)

# Convert to numeric values if needed (by stripping % sign if it's a string)
location_satisfaction_numeric = location_satisfaction.applymap(lambda x: float(x.strip('%')) if isinstance(x, str) else x)

# Plot a stacked bar chart
location_satisfaction_numeric.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(10, 6))

# Add labels and title
plt.title("Satisfaction with Remote Work by Work Location")
plt.xlabel("Work Location")
plt.ylabel("Percentage (%)")
plt.legend(title="Satisfaction Level")
plt.show()
```
